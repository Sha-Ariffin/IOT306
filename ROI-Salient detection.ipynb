```python
#Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
print("Libraries Imported")
```

```python
#Load Image
img = cv2.imread("people1.jpg")

# Convert BGR to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,6))
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#Apply Salient Detection
saliency = cv2.saliency.StaticSaliencySpectralResidual_create()
(success, saliencyMap) = saliency.computeSaliency(img)
saliencyMap = (saliencyMap * 255).astype("uint8")

plt.figure(figsize=(8,6))
plt.imshow(saliencyMap, cmap='gray')
plt.title("Saliency Map")
plt.axis("off")
plt.show()
```

```python
#Convert Saliency to ROI
thresh_value = 180
_, roi_mask = cv2.threshold(saliencyMap, thresh_value, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(8,6))
plt.imshow(roi_mask, cmap='gray')
plt.title("ROI Mask")
plt.axis("off")
plt.show()
```

```python
#Extract ROI from Original Image
roi = cv2.bitwise_and(img_rgb, img_rgb, mask=roi_mask)

plt.figure(figsize=(8,6))
plt.imshow(roi)
plt.title("Detected ROI")
plt.axis("off")
plt.show()
```

```python
#Draw Bounding Box
contours, _ = cv2.findContours(roi_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

output = img_rgb.copy()

for cnt in contours:
    x, y, w, h = cv2.boundingRect(cnt)
    
    if w*h > 500:   # ignore tiny noise
        cv2.rectangle(output, (x,y), (x+w,y+h), (255,0,0), 2)

plt.figure(figsize=(8,6))
plt.imshow(output)
plt.title("ROI Bounding Box")
plt.axis("off")
plt.show()
```
